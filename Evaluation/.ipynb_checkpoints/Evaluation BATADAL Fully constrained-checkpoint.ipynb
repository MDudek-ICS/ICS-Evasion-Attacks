```python
import sys
sys.path.append('../')
from Attacked_Model.autoencoder_BATADAL import load_AEED
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, roc_curve, auc, precision_score, recall_score
import pickle
from sklearn.preprocessing import MinMaxScaler
import numpy as np
att_data = pd.read_csv('../Data/BATADAL/attack_10_from_test_dataset.csv')

# define the column sets for the pandas dataframes
xset = [col for col in att_data.columns if col not in ['Unnamed: 0', 'DATETIME', 'ATT_FLAG']]
yset = ['ATT_FLAG']

autoencoder = load_AEED("../Attacked_Model/BATADAL/autoencoder.json", "../Attacked_Model/BATADAL/autoencoder.h5")
scaler = pickle.load(open("../Attacked_Model/BATADAL/scaler.p", "rb"))
with open("../Attacked_Model/BATADAL/theta") as f:
        theta = float(f.read())
import warnings
warnings.filterwarnings('ignore')
```

```python
def compute_scores(Y, Yhat):
    return [accuracy_score(Y, Yhat), f1_score(Y, Yhat), precision_score(Y, Yhat), recall_score(Y, Yhat)]#, fpr[0]]
```

# Evaluate Constrained attack over X dimension

```python
results_mean_replay = {new_list: 0 for new_list in ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9']}
results_mean_white = {new_list: 0 for new_list in ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9']}
results_mean_black = {new_list: 0 for new_list in ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9']}
sum_orig = 0
all_results_replay = pd.DataFrame(
                           columns= ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9'])
all_results_white = pd.DataFrame(
                           columns= ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9'])
all_results_black = pd.DataFrame(
                           columns= ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9'])
results_orig = pd.DataFrame(
                           columns=['accuracy', 'f1_score', 'precision', 'recall'])#,'fpr'])
for i in range(1,15):
    df_test_01 = pd.read_csv('../Data/BATADAL/attack_'+str(i)+'_from_test_dataset.csv', parse_dates = ['DATETIME'], dayfirst=True)
    X3 = pd.DataFrame(index=df_test_01.index, columns=xset,
                      data=scaler.transform(df_test_01[xset]))    
    Y3 = [1]*len(X3)
    Yhat3, _, _, _ = autoencoder.detect(X3, theta=theta, window=3, average=True)
    
    results_orig.loc[i] = compute_scores(Y3, Yhat3)
    sum_orig = sum_orig + results_orig.loc[i].at['recall']
    results_replay = {}
    results_white = {}
    results_black = {}
    for PLC in ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9']:
            
     
            replay = pd.read_csv('../Adversarial_Attacks/Replay_Attack/results/BATADAL/constrained_PLC/constrained_'+PLC+'_attack_'+str(i)+
                                        '.csv', dayfirst=True)
            white_box = pd.read_csv('../Adversarial_Attacks/Whitebox_Attack/results/BATADAL/max_constraints_fixed/whitebox_attack_'+str(i)+
                                        '_from_test_dataset'+PLC+'.csv', dayfirst=True)
            black_box = pd.read_csv('../Adversarial_Attacks/Black_Box_Attack/results/BATADAL/AE_PLC_constraint/new_constrained_plc_AE_attack_'+str(i)+
                                    '_from_test_dataset_'+PLC+'.csv', dayfirst=True)
           
            replay = pd.DataFrame(index=replay.index, columns=xset,
                      data=scaler.transform(replay[xset]))
            white_box = pd.DataFrame(index=white_box.index, columns=xset,
                      data=scaler.transform(white_box[xset]))
            black_box = pd.DataFrame(index=black_box.index, columns=xset,
                      data=scaler.transform(black_box[xset]))
            
            Y4 = [1]*len(replay)
            Y5 = [1]*len(white_box)
            Y6 = [1]*len(black_box)
            Yhat4, _, _, _ = autoencoder.detect(replay, theta=theta, window=3, average=True)
            Yhat5, _, _, _ = autoencoder.detect(white_box, theta=theta, window=3, average=True)
            Yhat6, _, _, _ = autoencoder.detect(black_box, theta=theta, window=3, average=True)
            
            results = pd.DataFrame(
                           columns=['accuracy', 'f1_score', 'precision', 'recall'])#,'fpr'])
            results.loc['replay'] = compute_scores(Y4, Yhat4)
            results.loc['white'] = compute_scores(Y5, Yhat5)
            results.loc['black'] = compute_scores(Y6, Yhat6)
            
            results_replay[PLC] = results.loc['replay'].at['recall']
            results_white[PLC] = results.loc['white'].at['recall']
            results_black[PLC] = results.loc['black'].at['recall']  
            
            results_mean_replay[PLC] = results_mean_replay[PLC] + results_replay[PLC]
            results_mean_white[PLC] = results_mean_white[PLC] + results_white[PLC]
            results_mean_black[PLC] = results_mean_black[PLC] + results_black[PLC]
    
    all_results_replay.loc[i] =  results_replay   
    all_results_white.loc[i] =  results_white
    all_results_black.loc[i] =  results_black
    
```

```python
results_attack = pd.DataFrame(columns=['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9'])
for j in ['PLC_1','PLC_2', 'PLC_3', 'PLC_4', 'PLC_5', 'PLC_6', 'PLC_7', 'PLC_8', 'PLC_9']:
    results_attack.loc['replay', j] = round(results_mean_replay[j]/14, 2)
    results_attack.loc['white', j] = round(results_mean_white[j]/14, 2)
    results_attack.loc['black', j] = round(results_mean_black[j]/14, 2)
```
